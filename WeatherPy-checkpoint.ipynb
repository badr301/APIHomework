```python
# Dependencies and Setup
import openweathermapy.core as owm
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from api_keys import api_key


#47b7c95ab160dff07b5ab75e1e49b4ee
#http://history.openweathermap.org/data/2.5/history/city?lat={lat}&lon={lon}&type=hour&start={start}&end={end}


# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy


settings = {"units": "metric", "appid": api_key}

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"

# Build partial query URL
query_url = f"{url}appid={api_key}&units={units}&q="
query_url
```

```python
#cities = ["Paris", "London", "Oslo", "Beijing"]
#cities = []
# set up lists to hold reponse info
lat = []
lon = []
temp = []
hum = []
win = []
cld = []

# Loop through the list of cities and perform a request for data on each
for city in cities:
    #response = requests.get(query_url + city).json()
    # Get current weather
    try:
        response = owm.get_current(city, **settings)
    except:
        print('Not Found')
    lat.append(response['coord']['lat'])
    temp.append(response['main']['temp'])
    lon.append(response['coord']['lon'])
    hum.append(response['main']['humidity'])
    win.append(response['wind']['speed'])
    cld.append(response['clouds']['all'])   
   
    
print(f"The latitude information received is: {lat}")
print(f"The temperature information received is: {temp}")
print(f"The latitude information received is: {lon}")
print(f"The temperature information received is: {hum}")
print(f"The latitude information received is: {win}")
print(f"The latitude information received is: {cld}")



```

```python
weather_dict = {
    "city": cities,
    "lat": lat,
    "Lng": lon,
    "temp": temp,
    "Humidity": hum,
    "Wind Speed": win,
    "Cloudiness": cld
    
}
weather_data = pd.DataFrame(weather_dict)
weather_data.head()



```
